```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
theta = [100, 102, 101, 105, 107, 106, 108, 110, 109, 111]
```

```python
data = pd.Series(theta)
```

```python
data
```

```python
beta = 0.9
v = [theta[0]]
```

```python
for t in range(1, len(theta)): 
    v_t = beta*v[t-1]+(1-beta)*theta[t]
    v.append(v_t)
ewma = pd.Series(v)    
```

```python
print(ewma)
```

```python
plt.figure(figsize=(10,5))
plt.plot(data, label='Original Data (theta_t)', marker='o')
plt.plot(ewma, label='EWMA Smoothed (v_t)', marker='x')
plt.title('EWMA Practical Example')
plt.xlabel('Time t')
plt.ylabel('Value')
plt.legend()
plt.show()
```
