# Dropout Model for Regression

```python
model_2 = Sequential()
model_2.add(Dense(128,input_dim=1,activation="relu"))
model_2.add(Dropout(0.2))
model_2.add(Dense(128,activation="relu"))
model_2.add(Dropout(0.2))
model_2.add(Dense(1,activation="linear"))
adam=Adam(learning_rate=0.01)
model_2.compile(loss='mse', optimizer=adam,metrics=['mse'])
drop_out_history=model_2.fit(X_train,Y_train,epochs=500, validation_data=(X_test,Y_test),verbose=False)
```

```python
# evaluate the model 
_,train_mse = model_2.evaluate(X_train,Y_train,verbose=0)
_,test_mse = model_2.evaluate(X_test,Y_test,verbose=0)
print('Train:{},Test:{}'.format(train_mse,test_mse))
```

```python
# in less dropout value it will overfit but with high dropout value it dosen't.
# if overfitting then increase the value of dropout(P)
# if underfitting then decrese the value of dropout 
# put dropout in last layer first then others 
# CNN 40% to 50% Dropout ratio
# ANN 10 - 50 % Dropout ratio
# RNN less than 50 % ratio
```
